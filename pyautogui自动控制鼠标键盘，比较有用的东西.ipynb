```python
import pyautogui,time
pyautogui.PAUSE=0.5
nameField=(487,796)
#pyautogui.press('end')
grayBlock=(460,379)
grayBlockColor=(227,227,227)
submitAnotherLink=(565,607)
formData=[{'name':'Alice','fear':'eavesdroppers','source':'wand','robocop':4,'comments':'Tell bob i said hi.'},{'name': 'Bob', 'fear': 'bees', 'source': 'amulet', 'robocop': 4, 'comments': 'null'},{'name': 'Carol', 'fear': 'puppets', 'source': 'crystal ball', 'robocop': 1, 'comments': 'Please take the puppets out of the break room.'},
{'name': 'Alex Murphy', 'fear': 'ED-209', 'source': 'money', 'robocop': 5, 'comments': 'Protect the innocent. Serve the public trust. Uphold the law.'}]

for person in formData:
    #Give the user a chance to kill the script.
    print('>>>5 SECOND PAUSE TO LET USER PRESS CTRL-C<<<')
    time.sleep(5)
    
    #Wait until the form page has loaded.
    while not pyautogui.pixelMatchesColor(grayBlock[0],grayBlock[1],grayBlockColor):
        time.sleep(0.5)
        pyautogui.press('home')
    print('entering %s info...'%(person['name']))
    pyautogui.click(nameField[0],nameField[1])
    #Fill out the name field
    pyautogui.typewrite(person['name']+'\t')
    #Fill out the Greast Fear field
    pyautogui.typewrite(person['fear']+'\t')
    #fill out the source of wizard power field.
    if person['source']=='wand':
        pyautogui.press(['down'])
        pyautogui.press('enter')
        pyautogui.press('\t')
    elif person['source']=='amulet':
        pyautogui.press(['down','down'])
        pyautogui.press('enter')
        pyautogui.press('\t')
    elif person['source']=='crystal ball':
        pyautogui.press('down')
        pyautogui.press('down')
        pyautogui.press('down')
        pyautogui.press('enter')
        pyautogui.press('\t')
    elif person['source']=='money':
        time.sleep(1)
        pyautogui.press('down')
        pyautogui.press('down')
        pyautogui.press('down')
        pyautogui.press('down')
        pyautogui.press('enter')
        pyautogui.press('\t')
    #fill out the RoboCop field.
    if person['robocop']==1:
        pyautogui.typewrite(['left','right','\t'])
    elif person['robocop']==2:
        pyautogui.typewrite(['right','\t'])
    elif person['robocop']==3:
        pyautogui.typewrite(['right','right','\t'])
    elif person['robocop']==4:
        print(person['name'])
        pyautogui.typewrite(['right','right','right','\t'])
    elif person['robocop']==5:
        pyautogui.typewrite(['right','right','right','right','\t'])
    #fill out the additional comment field
    #pyautogui.typewrite(['\t','backspace'])
    pyautogui.typewrite(person['comments']+'\t')
    
    #click submit
    pyautogui.press('enter')
    #wait until form page has loaded.
    print('clicked submit.')
    time.sleep(5)
    #click the submit another response link.
    pyautogui.click(submitAnotherLink[0],submitAnotherLink[1])
```
