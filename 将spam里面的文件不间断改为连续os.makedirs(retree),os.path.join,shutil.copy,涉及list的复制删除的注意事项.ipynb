```python
import os,shutil
#进入当前文件夹路径
filepath=input('输入spam所在的文件夹路径：')
os.chdir(filepath)
#将spamxxx放进一个list中,放一个删除一个
#print(os.listdir())
emlist=[]
halist=[]
'''在halist下创建一个insertgaps文件夹'''
insertfilepath=os.path.join(filepath,'insertgaps')
if os.path.exists(insertfilepath):
    shutil.rmtree(insertfilepath)
os.makedirs(insertfilepath)
#如果done文件夹存在，删除，如果不存在创建
copyfilepath=os.path.join(filepath,'done')
if os.path.exists(copyfilepath):
    shutil.rmtree(copyfilepath)
os.makedirs(copyfilepath)
orglist=os.listdir()
#筛选出所有以spam开头的文件
orglistcopy=orglist[:]
for i in orglistcopy:
    if i.startswith('spam')==False:
        orglist.remove(i)
    elif i.startswith('spam')==True: 
        shutil.copy(i,insertfilepath)
        '''在insert文件夹中，先把所有的spam放进去'''
i=1
maxspam=0
while len(orglist)>0:
    #print(orglist)
    begwd='spam'+str(i).zfill(3)
#flag为判断标签，全部不匹配才能确定不在序列中
    flag=False
    for file in orglist:
        flag=flag|file.startswith(begwd)
        if file.startswith(begwd):
            #按顺序放入list中，索引便能指示出区别
            halist.append(file)
            #找到就删除提高算法效率
            orglist.remove(file)
            break
#此步应该还可以优化我写不出来了
    if flag==False:
        emlist.append(begwd)
    i=i+1
    maxspam=i
#下一步修改halist里面的名字从001开始计算,并在当前文件的子目录中生成一个done的文件夹来寄放
for i in halist:
    #构造出每一个位置的标准文件名
    shouldfilename='spam'+str(halist.index(i)+1).zfill(3)+i[7:]
    #如果不匹配，修改名字进行复制移动
    if shouldfilename!=i:
        copyfile=os.path.join(copyfilepath,shouldfilename)
        shutil.copy(i,copyfile)
    else:
    #如果匹配，不改名字进行复制移动
        copyfile=os.path.join(copyfilepath,i)
        shutil.copy(i,copyfile)
'''输入你想insert进去的文件名，注意当你insert当中间没有间隙的时候得加到最后了'''
needinsertfile=input("输入你想插入的文件名的绝对路径：")
if len(emlist)>0:
    for i in emlist:
        #调整需要插入的文件的名字
        shouldfilename=i+os.path.basename(needinsertfile)
        insertfilename=os.path.join(insertfilepath,shouldfilename)
        shutil.copy(needinsertfile,insertfilename)
        emlist.remove(i)
        break
else:
    shouldfilename='spam'+str(maxspam).zfill(3)+os.path.basename(needinsertfile)
    insertfilename=os.path.join(insertfilepath,shouldfilename)
    shutil.copy(needinsertfile,insertfilename)

        

        
```
