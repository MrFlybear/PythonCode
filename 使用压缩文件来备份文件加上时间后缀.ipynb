```python
#漂亮的程序
import zipfile,os,datetime
#print(os.getcwd())
folder=input("请输入你想备份的文件夹的绝对路径，备份的zip文件将在该文件夹内生成并添加备份时间后缀：")
def backupToZip(folder):
    date_now=datetime.datetime.now()
    datetail=str(date_now.year).zfill(4)+str(date_now.month).zfill(2)+str(date_now.day).zfill(2)+str(date_now.hour).zfill(2)+str(date_now.minute).zfill(2)
    #加上时间后缀的zip文件名称
    zipFilename=os.path.basename(folder)+'_'+datetail+'.zip'
    #压缩文件的绝对路径
    zipFilename=os.path.join(folder,zipFilename)
    print(zipFilename)
    backupzip=zipfile.ZipFile(zipFilename,'w')
    #调整当前工作路径
    dirfolder=os.path.dirname(folder)
    os.chdir(dirfolder)
    for foldername,subfolders,filenames in os.walk(os.path.basename(folder)):
        for filename in filenames:
            if filename.endswith('.zip'):
                continue
            print(filename)
            backupzip.write(os.path.join(foldername,filename))
    backupzip.close()
backupToZip(folder)
            
    
```
