```python
import zipfile,os
print(os.getcwd())
#生成一个zip文件，然后把另一个文件夹的东西写进去，居然是先生成zip，可再抽象懒得弄了
newZip=zipfile.ZipFile(r'xiang.zip','w')
os.chdir(r'C:\Users\xxcmt\Desktop')
testabspath=os.path.abspath(r'test')
#调整绝对路径并把里面的basename作为walk对象，压缩时将不产生父级文件夹
for foldername,subfolders,filenames in os.walk(os.path.basename(testabspath)):
    for filename in filenames:
        print(os.path.join(foldername,filename))
        newZip.write(os.path.join(foldername,filename))
newZip.close()
```
