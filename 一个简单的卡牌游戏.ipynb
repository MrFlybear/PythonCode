```python
import random
class Card:
    suit_name=['fangpian','meihua','hongxin','heitao']
    rank_name=dict()
    rank_name[1]='Ace'
    for i in range(2,11):
        rank_name[i]=i
    rank_name[11]='jack'
    rank_name[12]='Queen'
    rank_name[13]='king'
    def __init__(self,suit=0,rank=13):
        '''
        suit:0 Clubs(方块),1
        '''
        self.suit=suit
        self.rank=rank
    def __str__(self):
        return '%s  %s' %(Card.suit_name[self.suit],
                           Card.rank_name[self.rank])
    def __lt__(self,other):
        t1=self.suit,self.rank
        t2=other.suit,other.rank
        return t1<t2
class Deck:
    def __init__(self):
        self.cards=[]
        for suit in range(4):
            for rank in range(1,14):
                card=Card(suit,rank)
                self.cards.append(card)
    def __str__(self):
        res=[]
        for card in self.cards:
            res.append(str(card))
        return '\n'.join(res)
    
    def pop_card(self):
        return self.cards.pop
    
    def add_card(self,card):
        return self.cards.append(card)
    
    def shuffle(self):
        random.shuffle(self.cards)
        
    def sort(self):
        return self.cards.sort()
    
    def move_cards(self,hand,num):
        for i in range(num):
            hand.add_card(self.pop_card())
    deck=Deck()
    shuffle(deck)
    print(deck)
    sort(deck)
    print(deck)

class Hand(Deck):
    def __init__(self,label=''):
        self.cards=[]
        self.label=label
    
        
```
