```python
#蛋蛋的自己抓,编码真恶心
import requests,bs4,pprint,time,os
from selenium import webdriver
import smtplib,sys,chardet,itchat

#用浏览器加载js之后的真response，名为htmlname存硬盘上
def BrowserResponse(url,htmlname):
    os.chdir(r'C:\Users\xxcmt\Desktop\文档\automate_online-materials')
    browser = webdriver.Firefox()
    browser.get(url)
    time.sleep(2)
    html_text=browser.page_source
    browser.quit()
    with open(htmlname,'w',encoding='utf-8') as f:
        f.write(html_text)

#打开已存的htmlname,返回时间-天气dict
def shenzhenWeather(htmlname):
    os.chdir(r'C:\Users\xxcmt\Desktop\文档\automate_online-materials')
    #res返回的是response
    with open(htmlname,'r',encoding='utf-8') as res:
        soup=bs4.BeautifulSoup(res,'html')
    rainElems=soup.select(' .wpic > div  big')
    timeElems=soup.select('div.time:nth-child(1) > em')
    time_list=[]
    weather_list=[]
    for i in timeElems:
        time_list.append(i.get_text())
    for i in rainElems:
        weather_list.append(i.get('title'))
    timeweather_dict=dict(zip(time_list,weather_list))
    return timeweather_dict

#判断带不带伞，并返回可发送的str
def isUmbrella(todayWeather):
    strRainTime=''
    for key,value in todayWeather.items():
        if '雨' in value:
            strRainTime=strRainTime+key+' '
        else:
            continue
    if strRainTime!='':                
        str2send='亲爱的懒猪，有雨,记得带伞'
        #str2send='亲爱的懒猪，今日天气是：\n%s\n%s 有雨,记得带雨'%(pprint.pformat(todayWeather),strRainTime)
    else:
        str2send='今日无雨'
    #mail2send='Subject:todayrain?\n\n'+str2send
    mail2send=str2send
    return mail2send.encode('utf-8')

def sendmail(content,reciver):
    smtpObj=smtplib.SMTP('smtp.qq.com',587)
    smtpObj.ehlo()
    smtpObj.starttls()
    #要开启SMTP获得授权码
    smtpObj.login('1036433209@qq.com','ztmgnjogoyjhbedh')
    smtpObj.sendmail('1036433209@qq.com',reciver,content)
    smtpObj.quit()

def sendIchat(content):
    itchat.auto_login(hotReload=True)
    users=itchat.search_friends(name='cmt')
    userName=users[0]['UserName']
    itchat.send(content.decode('utf-8'),toUserName=userName)
    print('微信已发')
    
BrowserResponse("http://www.weather.com.cn/weather1d/101280606.shtml#input",'lgweather.html')
todayWeather=shenzhenWeather('lgweather.html')
content=isUmbrella(todayWeather)
print(content.decode('utf-8'))
#sendmail(content,'1036433209@qq.com')
sendIchat(content)

    

   
    
```
