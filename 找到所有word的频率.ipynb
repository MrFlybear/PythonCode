```python
#去掉空格和单词,基本满足要求，不搞了
import string
import random
fp=open('C:\Alice.txt')
symbol=string.punctuation+'‘’“”'+string.whitespace
hisgram=dict()
print(hisgram)
for line in fp:
    for word in line.split():
        word=word.strip(symbol)
        if word not in hisgram:
            hisgram[word]=1
        else:
            hisgram[word]=hisgram[word]+1
#for key in hisgram:
#    print(key,hisgram[key])
t=[]
for word,freq in hisgram.items():
    t.extend([word]*freq)
m=len(t)
for i in range(100):
    print(t[random.randint(0,m)])


```
